#Problem Statement

FICO scores are continuous numeric values

The ML architecture requires categorical inputs

Hard-coded score ranges:

Are subjective

Do not generalize well to future datasets

#Objective

Create an optimal, data-driven bucketing strategy
that converts FICO scores into risk-representative categories

#Why Bucketing Matters

Incorrect Bucketing Can Lead To:

Mixing high-risk and low-risk borrowers

Unstable PD estimates

Poor model performance

Good Bucketing Should:

Group borrowers with similar default behavior

Preserve risk separation

Work consistently on future portfolios

#What is Quantization?

Converting a continuous variable into discrete buckets

Assigning one representative value (or label) per bucket

In This Use Case:

Input: FICO score (300–850)

Output: Credit rating category

Lower rating = better credit quality

#Approaches
Mean Squared Error (MSE)
Log-Likelihood Maximization

Approach 1: Mean Squared Error (MSE)

How It Works

Groups FICO scores to minimize numerical error

Treats scores as a signal approximation problem

Advantages

Simple

Preserves numeric structure

Limitations

Does not directly model default behavior

Ignores concentration of risk within buckets

Not optimal for PD-focused modeling

Approach 2: Log-Likelihood (Chosen Approach)

How It Works

Uses observed defaults to guide bucket formation

Each bucket has:

Total borrowers (n)

Defaults (k)

PD = k / n

Objective

Maximize the likelihood of observing actual defaults

Create buckets that are risk-homogeneous
